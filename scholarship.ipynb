```python
import requests
from bs4 import BeautifulSoup
import json
import re

```

```python
#url  = "https://www.vidhyaa.in/blog/scholarship-for-indian-students-after-12th"
def fetch_html(url):
    response = requests.get(url)
    return response.text

def parse_html(html):
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

```python
def extract_scholarships(soup):
    scholarships = []
    scholarship_sections = soup.select('.raman-base li a')  # Update CSS selector as needed

    for section in scholarship_sections:
        title = section.get_text().strip()
        url = section.get('href')

        # Fetch each scholarship page
        scholarship_html = fetch_html(url)
        scholarship_soup = BeautifulSoup(scholarship_html, 'html.parser')

        # Extract the relevant sections from each scholarship page
        full_content = scholarship_soup.select_one('.raman-base')  # Update CSS selector for content
        if full_content:
            text = full_content.get_text(separator="\n")
            scholarships.append({
                "title": title,
                "url": url,
                "content": text
            })

    return scholarships
```

```python
def convert_to_json(scholarships):
    return json.dumps(scholarships, indent=4)
```

```python
def parse_scholarship_details(content):
    # Helper function to find a section using regex
    def find_section(pattern, text):
        match = re.search(pattern, text, re.DOTALL)
        return match.group(1).strip() if match else ''

    # Define regex patterns for each section
    eligibility_pattern = r"Eligibility Criteria\s*(.*?)\n(?=Raman Kant Munjal \(RKM\) Scholarship 2023 Rewards)"
    rewards_pattern = r"Scholarship 2023 Rewards\s*(.*?)\n(?=Raman Kant Munjal Scholarship 2023 Online Form)"
    application_process_pattern = r"Scholarship 2023 Online Form\s*(.*?)\n(?=Documents To Be Attached)"
    documents_pattern = r"Attached For Raman Kant Munjal Scholarship\s*(.*?)\n(?=Selection Criteria)"
    selection_criteria_pattern = r"Selection Criteria Of Raman Kant Munjal \(RKM\) Scholarship\s*(.*?)\n(?=Check Some Other List Of Scholarship)"
    other_scholarships_pattern = r"List Of Scholarship For Scholarships For Indian Students after 12th\s*(.*?)\n(?=Raman Kant Munjal Scholarship 2023 Last Date)"
    last_date_pattern = r"Scholarship 2023 Last Date\s*(.*?)\n(?=Terms and condition)"
    terms_and_conditions_pattern = r"Terms and condition\s*(.*)"

    # Extract each section
    eligibility = find_section(eligibility_pattern, content)
    rewards = find_section(rewards_pattern, content)
    application_process = find_section(application_process_pattern, content)
    required_documents = find_section(documents_pattern, content)
    selection_criteria = find_section(selection_criteria_pattern, content)
    other_scholarships = find_section(other_scholarships_pattern, content)
    last_date = find_section(last_date_pattern, content)
    terms_and_conditions = find_section(terms_and_conditions_pattern, content)

    return {
        "eligibilityCriteria": eligibility,
        "rewards": rewards,
        "applicationProcess": application_process,
        "requiredDocuments": required_documents,
        "selectionCriteria": selection_criteria,
        "otherScholarships": other_scholarships,
        "applicationDeadline": last_date,
        "termsAndConditions": terms_and_conditions
    }
 # Your scholarship content string


if __name__ == "__main__":
    main()
def main():
    url = "https://www.vidhyaa.in/blog/scholarship-for-indian-students-after-12th"
    html_data = fetch_html(url)
    soup = parse_html(html_data)
    scholarships = extract_scholarships(soup)
    json_data = convert_to_json(scholarships[0])
    scholarship_content =json_data 
    parsed_scholarship = parse_scholarship_details(scholarship_content)
    print(json.dumps(parsed_scholarship, indent=4))
    # print(json_data)

# if __name__ == "__main__":
#     main()
# def main():
#     url = "https://www.vidhyaa.in/blog/scholarship-for-indian-students-after-12th"
#     html_data = fetch_html(url)
#     soup = parse_html(html_data)
#     scholarships = extract_scholarships(soup)
#     json_data = convert_to_json(scholarships)
#     print(json_data)

# if __name__ == "__main__":
#     main()
```
